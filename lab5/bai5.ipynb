```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

from sklearn import datasets, tree, metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import ConfusionMatrixDisplay
from mlxtend.plotting import plot_confusion_matrix

from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
```

```python
digits = datasets.load_digits()
```

```python
_, axes = plt.subplots(1, 10)
images_and_labels = list(zip(digits.images, digits.target))
for ax, (image, label) in zip(axes, images_and_labels[:10]):
    ax.set_axis_off()
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation ='nearest')
    ax.set_title('Traing: %i' % label)
plt.show()
```

```python
n_samples = len(digits.images)
data = digits.images.reshape((n_samples, -1))
```

```python
Y = digits.target
```

```python
algorithm = KMeans(n_clusters=10, init='random')
algorithm.fit(data)
labels = algorithm.labels_
centroids = algorithm.cluster_centers_
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, random_state=42)
data_tsne = tsne.fit_transform(data)
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)

ax1.scatter(data_tsne[:,0],data_tsne[:,1], c = labels, cmap = "jet", edgecolor = "None", alpha=0.35)
ax1.set_title('k-means clustering plot')
ax2.scatter(data_tsne[:,0],data_tsne[:,1], c = Y , cmap = "jet", edgecolor = "None", alpha=0.35)
ax2.set_title('Actual clusters')
```
